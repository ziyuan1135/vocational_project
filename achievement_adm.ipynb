```python
import pandas as pd
import numpy as np
import mysql.connector
```

```python
def extractdepCampusNote(strDepartment):
    dep, campus, note, isNote = '', '0', '0', True
    if(strDepartment.find('校區') > -1):
        campus = ''
        isNote = False
    for i in range(len(strDepartment)):
        if(strDepartment[i] == '(' or strDepartment[i] == '（'):
            if(isNote):
                note = ""
            for j in range(i+1, len(strDepartment)):
                if(strDepartment[j] == ')' or strDepartment[j] == '）'):
                    break
                if(isNote):
                    note += strDepartment[j]
                else:
                    campus += strDepartment[j]
            break
        elif(strDepartment[i] == ' ' or strDepartment[i] == '　'):
            continue
        dep += strDepartment[i]
    return list((dep, campus, note))

def validSchool(campusList, addressList):
    if(len(campusList) != len(addressList)):
        return False
    if(len(campusList) == 1):
        if(campusList[0] == '0' and addressList[0] != '0'):
            return True
        else:
            return False
    else:
        for i in range(len(campusList)):
            if(campusList[i] == '0' or addressList[i] == '0'):
                return False
        return True
```

### 統整112技優甄選與保送Excel

```python
dfAudition = pd.read_excel('./文件集/112技優甄選.xls')
dfRecommend = pd.read_excel('./文件集/112技優保送.xls')
dfAudition = dfAudition[['學校代碼', '學校名稱', '系科組學程名稱', '類別代碼']].set_axis(['學校ID', '校名', '科系', '類別名稱'], axis='columns')
dfRecommend = dfRecommend[['學校代碼', '學校名稱', '系科組學程名稱', '類別代碼']].set_axis(['學校ID', '校名', '科系', '類別名稱'], axis='columns')
```

```python
keys = ['科系', '校區', '備註', '類別ID', '類別名稱']
dicDataAud = {'科系':[], '校區':[], '備註':[], '類別名稱':[], '類別ID':[]} 
dicDataRec = {'科系':[], '校區':[], '備註':[], '類別名稱':[], '類別ID':[]} 
for i in range(len(dfAudition)):
    tempList = extractdepCampusNote(dfAudition.iloc[i]['科系'])
    tempList.extend([dfAudition.iloc[i]['類別名稱'][:2], dfAudition.iloc[i]['類別名稱'][2:]])
    for j in range(len(keys)):
        dicDataAud[keys[j]].append(tempList[j])

for i in range(len(dfRecommend)):
    tempList = extractdepCampusNote(dfRecommend.iloc[i]['科系'])
    tempList.extend([dfRecommend.iloc[i]['類別名稱'][:2], dfRecommend.iloc[i]['類別名稱'][2:]])
    for j in range(len(keys)):
        dicDataRec[keys[j]].append(tempList[j])

for key in keys:
    dfAudition[key] = dicDataAud[key]
    dfRecommend[key] = dicDataRec[key]
dfAudition = dfAudition.iloc[:, [0, 1, 2, 5, 4, 6, 3]].sort_values(by=['學校ID', '校區', '科系'], ignore_index=True)
dfRecommend = dfRecommend.iloc[:, [0, 1, 2, 5, 4, 6, 3]].sort_values(by=['學校ID', '校區', '科系'], ignore_index=True)
```

```python
dfAudition.to_excel("./文件集/112技優甄選v2.xlsx", index=False)
dfRecommend.to_excel("./文件集/112技優保送v2.xlsx", index=False)
```
