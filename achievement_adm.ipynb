```python
import pandas as pd
import numpy as np
import mysql.connector
```

### 建立技優DataToDB

```python
dfAchievement = pd.read_excel("./文件集/技優類別統整.xlsx")
dfAchievement['類別ID'] = dfAchievement['類別ID'].astype('str')
```

```python
data = []
for i in range(len(dfAchievement)):
    data.append((dfAchievement.iloc[i][0], dfAchievement.iloc[i][1]))
data
```

```python
conn = mysql.connector.connect(user='root', password='root', host='localhost', database='四技二專')
cursor = conn.cursor()
sql = "insert into 技優 values(%s, %s, null)"
cursor.executemany(sql, data)
conn.commit()
```

```python
cursor.close()
conn.close()
```

### 彙整112技優類別Excel

```python
dfRec = pd.read_excel("./文件集/112技優甄選.xls")
dfConveyed = pd.read_excel('./文件集/112技優保送.xls')
print(dfRec['類別代碼'].unique(), len(dfRec['類別代碼'].unique()))
dfConveyed['類別代碼'].unique(), len(dfConveyed['類別代碼'].unique())
```

```python
temp = np.append(np.sort(dfRec['類別代碼'].unique()), np.sort(dfConveyed['類別代碼'].unique()))
classes = np.unique(temp)
classes, len(classes), type(classes)
```

```python
achievement_adm_data = []
for group in classes:
    achievement_adm_data.append([group[:2], group[2:]])
achievement_adm_data
```

```python
dfAchievement = pd.DataFrame(achievement_adm_data, columns=['類別ID', '類別名稱'])
dfAchievement = dfAchievement.drop(17).reset_index(drop=True)
dfAchievement
```

```python
dfAchievement.to_excel("./文件集/技優類別統整.xlsx", index=False)
```
