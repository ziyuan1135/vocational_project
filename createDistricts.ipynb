```python
import pandas as pd
import re
table = pd.read_html("http://www.i-write.idv.tw/life/info/postalcode/postalcode8.html")
```

```python
dfTable = table[3]
dfTable
```

```python
dicCitys = {}
for i in range(len(dfTable)):
    if(not bool(re.search('[\d]', dfTable.iloc[i][0]))):
        city = dfTable.iloc[i][0]
        dicCitys[city] = []
        continue
    for j in range(7):
        content, district, zoneNum = dfTable.iloc[i][j], "", ""
        if(content != content):
            break
        for k in range(len(content)):
            if(content[k] < '0' or content[k] > '9'):
                if(content[k] != " "): district += content[k]
            else:
                zoneNum += content[k]
        dicCitys[city].append([district, zoneNum])
```

```python
dicCitys['桃園市'] = dicCitys.pop('桃園縣')
dicCitys
```

```python
dicTW = {'北部' : ['台北市', '新北市', '基隆市', '桃園市', '新竹市', '新竹縣', '宜蘭縣'], 
         '中部' : ['苗栗縣', '台中市', '彰化縣', '南投縣', '雲林縣'], 
         '南部' : ['嘉義市', '嘉義縣', '台南市', '高雄市', '屏東縣', '澎湖縣'],
         '東部與其他' : ['花蓮縣', '台東縣', '金門縣', '連江縣', "南海諸島"]
        } 
data = []
for area in dicTW.keys():
    for city in dicTW[area]:
        for  district in dicCitys[city]:
            data.append([area, city, district[0], district[1]])
```

```python
dfCityIntegration = pd.DataFrame(data, columns=['地區', '縣市', '行政區名稱', '區號'])
dfCityIntegration
```

```python
dfCityIntegration.to_excel("./文件集/台灣縣市統整.xlsx", index=False)
```
