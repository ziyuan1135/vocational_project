### 驗證學校統整v3 and v4

```python
dfSchoolOld = pd.read_excel("./文件集/學校統整V3.xlsx")
dfSchoolNew = pd.read_excel("./文件集/學校統整v4.xlsx")
dfDepartment = pd.read_excel("./文件集/大學科系代碼v2.xlsx")
```

```python
def checkDepartment(dfSchoolNew):
    for department in dfSchoolNew['科系名稱']:
        for char in department:
            if(char == "(" or char == "（" or char == ")" or char == "）"):
                return False
    return True

def getDepartmentWithComment(dfSchoolOld):
    indexes = []
    for i in range(len(dfSchoolOld)):
        for char in dfSchoolOld.iloc[i]['科系名稱']:
            if(char == "(" or char == "（" or char == ")" or char == "）"):
                indexes.append(i)
                break
    return indexes
```

```python
print(checkDepartment(dfSchoolNew))
commentsIndexes = getDepartmentWithComment(dfSchoolOld)
mask = dfSchoolNew['備註'] != "0"
print(all(dfSchoolNew['備註'][mask].index == commentsIndexes))
```

### 修改學校統整v3.xlsx

```python
dfSchool = pd.read_excel("./文件集/學校統整V3.xlsx")
dfDepartment = pd.read_excel("./文件集/大學科系代碼(105).xls")
dfDepartment.columns = dfDepartment.loc[0]
dropIndex = [0]
dropIndex.extend(list(range(2300, 2310)))
dfDepartment = dfDepartment.drop(dropIndex).reset_index(drop=True)
```

```python
comments = []
departments = []
for departmentName in dfSchool['科系名稱']:
    comment = ""
    department = ""
    parentheses = False
    for i in range(len(departmentName)):
        if(parentheses):
            if(departmentName[i] != ')' and departmentName[i] != '）'):
                comment += departmentName[i]
            else:
                parentheses = False
        else:
            if(departmentName[i] == '(' or departmentName[i] == '（'):
                parentheses = True
            elif(departmentName[i] == ' ' or departmentName[i] == '　'):
                continue
            else:
                department += departmentName[i]
    comments.append(comment)
    departments.append(department)
```

```python
dfSchool['科系名稱'] = departments
dfSchool.insert(3, '備註', comments)
dfSchool['備註'].replace("", "0", inplace=True)
```

```python
dfSchool.to_excel("./文件集/學校統整v4.xlsx", index=False)
dfDepartment.to_excel("./文件集/大學科系代碼v2.xlsx", index=False)
```
