```python
import pandas as pd
```

```python
dfSchool = pd.read_excel("./文件集/學校統整V3.xlsx")
dfDepartment = pd.read_excel("./文件集/大學科系代碼(105).xls")
dfDepartment.columns = dfDepartment.loc[0]
dropIndex = [0]
dropIndex.extend(list(range(2300, 2310)))
dfDepartment = dfDepartment.drop(dropIndex).reset_index(drop=True)
```

```python
comments = []
departments = []
for departmentName in dfSchool['科系名稱']:
    comment = ""
    department = ""
    parentheses = False
    for i in range(len(departmentName)):
        if(parentheses):
            if(departmentName[i] != ')' and departmentName[i] != '）'):
                comment += departmentName[i]
            else:
                parentheses = False
        else:
            if(departmentName[i] == '(' or departmentName[i] == '（'):
                parentheses = True
            elif(departmentName[i] == ' ' or departmentName[i] == '　'):
                continue
            else:
                department += departmentName[i]
    comments.append(comment)
    departments.append(department)
```
